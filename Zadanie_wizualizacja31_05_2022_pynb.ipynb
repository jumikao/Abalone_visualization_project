```python
plt.figure(figsize=(10,8))

#liczenie korelacji między zmiennymi - count_corr
count_corr=uchowce100.corr()            

#heatmap
sns.heatmap(count_corr)

#tytuł
plt.title("Korelacja cech osobników", fontsize=20)


# Znak współczynnika korelacji informuje nas o kierunku korelacji, natomiast jego
# bezwzględna wartość o sile związku. Mamy oczywiście równość rxy = ryx. Jeśli rxy = 0,
# oznacza to zupełny brak związku korelacyjnego między badanymi zmiennymi X i Y
# (przypadek  na rysunku). Im wartość bezwzględna współczynnika korelacji jest bliższy
# jedności, tym zależność korelacyjna między zmiennymi jest silniejsza. Gdy rxy = |1|, to
# zależność korelacyjna przechodzi w zależność funkcyjną (funkcja liniowa).

# W analizie statystycznej zwykle przyjmuje się następującą skalę:
# rxy = 0 zmienne nie są skorelowane
# 0 < rxy < 0,1 korelacja nikła
# 0,1 < rxy < 0,3 korelacja słaba
# 0,3 < rxy < 0,5 korelacja przeciętna
# 0,5 < rxy < 0,7 korelacja wysoka
# 0,7 < rxy < 0,9 korelacja bardzo wysoka
# 0,9 < rxy < 1 korelacja prawie pełna.

#WNIOSKI
#Liczba pierścieni posiada bardzo wysoką korelację ze wszystkimi innymi cechami uchowców.
#Występuje wysoka korelacja pomiędzy wagą (ogólną, skorupy, wnętrzności oraz szkieletu wewnętrznego)
#, a wzrostem osobników.
#Szerokość i dlugość osobników sa nisko skorelowane ze sobą.
```
